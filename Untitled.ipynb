```python
import numpy as np
import matplotlib.pyplot as plt
from lmfit import Model
import sympy as sp

# Define the symbols and equation using SymPy
equation = "a * Meth + c"

# Generate a callable function using lambdify
model_callable = sp.lambdify(("Meth", "a", "c"), equation)

# Define the data
x_data = np.array([0, 0.3, 0.6, 0.9])
y_data = np.array([1,2,3, 4])

# Create a Model instance using the callable function
model = Model(model_callable, independent_vars=['Meth'])

# Initial guess for the parameters
params = model.make_params(a=1, c=0)

dev_var_dict = {"Meth": x_data}
print(dev_var_dict, "sfvsfdsdv")

# Fit the model to the data
result = model.fit(y_data, params, **dev_var_dict)

# Print the fit report
print(result.fit_report())

# Extract the fitted data
fitted_y = result.best_fit

# Plot the original data and the fitted line
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, fitted_y, label='Fitted Line', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Fit: y = a*x + c')
plt.show()
```
